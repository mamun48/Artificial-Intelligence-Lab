 #### Import the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Variables

```python
w = 4
b = 7
x = np.arange(-10,10,0.1,dtype=np.float64)
figure_size = (13,13)
```

### Function : y = wx + b

```python
y = w * x + b

# plt.figure(figsize=figure_size,dpi = 100)
plt.title('y = wx + y')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y)
plt.show()
```

### Function : y = x^2

```python
y = x * x

#plt.figure(figsize=figure_size,dpi = 100)
plt.title('y = x^2')
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

#### Function: y = (e^x - e^(-x)) / (e^x + e^(-x))

```python
y = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

plt.title('y = (e^x - e^(-x)) / (e^x + e^(-x))')
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
